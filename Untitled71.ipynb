```python
# Assuming 'species_column' is the name of the column with the repeated species names
data['species'] = data['species'].str.split('Iris-').str[1]

# Select only numeric columns for mean calculation
numeric_data = data.select_dtypes(include=['number'])
data_filled_mean = numeric_data.fillna(numeric_data.mean())
print(data_filled_mean)
```

```python
x= data[['sepal_length','sepal_width','petal_length','petal_width']]
y=data['species']
```

```python
x
```

```python
y
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
svm_model = SVC(kernel='linear')  # You can choose different kernels like 'rbf', 'poly', etc.
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('Classification Report:')
print(classification_report(y_test, y_pred))


```

performance of Cross Validation

```python
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.metrics import precision_score, recall_score, f1_score, make_scorer

# Perform cross-validation
scores = cross_val_score(svm_model, x, y, cv=5)  # 5-fold cross-validation
print(f'Cross-validation accuracy scores: {scores}')
print(f'Mean cross-validation accuracy: {scores.mean():.2f}')

# Evaluate additional metrics using cross-validation
scoring = {
    'accuracy': 'accuracy',
    'precision_macro': 'precision_macro',
    'recall_macro': 'recall_macro',
    'f1_macro': 'f1_macro'
}
cv_results = cross_validate(svm_model, x, y, cv=5, scoring=scoring)
print(f'Cross-validation precision: {cv_results["test_precision_macro"].mean():.2f}')
print(f'Cross-validation recall: {cv_results["test_recall_macro"].mean():.2f}')
print(f'Cross-validation F1-score: {cv_results["test_f1_macro"].mean():.2f}')
```

plotting of the accuracy

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(scores) + 1), scores, marker='o', linestyle='--', color='b')
plt.title('Cross-Validation Accuracy')
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])  # Set the y-axis limits to better visualize accuracy
plt.xticks(range(1, len(scores) + 1))
plt.grid(True)
plt.show()
```
